```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "Pymaceuticals_mouse_drug_data.csv"
clinical_trial_data_to_load = "Pymaceuticals_clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
Mouse_Data = pd.read_csv(mouse_drug_data_to_load)
Clinical_Data = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
combined_mouse_clinical_df = pd.merge(Mouse_Data, Clinical_Data,
                                 how='outer', on='Mouse ID')
```

```python
# Display the data table for preview

combined_mouse_clinical_df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
Mean_Tumor_Volume_data = combined_mouse_clinical_df.groupby(["Drug","Timepoint"])["Tumor Volume (mm3)"].mean()
# Convert to DataFrame
Mean_Tumor_Volume_df = pd.DataFrame(Mean_Tumor_Volume_data)
Mean_Tumor_Volume_df_Index = Mean_Tumor_Volume_df.reset_index(level=None, drop=False)
# Preview DataFrame
Mean_Tumor_Volume_df_Index.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
SEM_Tumor_Volume_data = combined_mouse_clinical_df.groupby(["Drug","Timepoint"])["Tumor Volume (mm3)"].sem()
# Convert to DataFrame
SEM_Tumor_Volume_df = pd.DataFrame(SEM_Tumor_Volume_data)
# Preview DataFrame
SEM_reset_df = SEM_Tumor_Volume_df.reset_index(level=None, drop=False)

SEM_reset_df
```

```python
# Minor Data Munging to Re-Format the Data Frames
Minor_Data_Reformatted_df = Mean_Tumor_Volume_df.pivot_table(values= 'Tumor Volume (mm3)', index='Timepoint', columns='Drug' )
# Preview that Reformatting worked
Minor_Data_Reformatted_df.head()
```

```python
# Generate the Plot (with Error Bars)
Capomulin_Error_Bar = SEM_reset_df.loc[SEM_reset_df["Drug"] == "Capomulin", "Tumor Volume (mm3)"]
Infubinol_Error_Bar = SEM_reset_df.loc[SEM_reset_df["Drug"] == "Infubinol", "Tumor Volume (mm3)"]
Ketapril_Error_Bar = SEM_reset_df.loc[SEM_reset_df["Drug"] == "Ketapril", "Tumor Volume (mm3)"]
Placebo_Error_Bar = SEM_reset_df.loc[SEM_reset_df["Drug"] == "Placebo", "Tumor Volume (mm3)"]

#set x axis
x_axis = np.arange(0,50,5)

plt.errorbar(x_axis, Minor_Data_Reformatted_df["Capomulin"] , yerr= Capomulin_Error_Bar, label= "Capomulin", marker= "o", color="red", linestyle='--')
plt.errorbar(x_axis, Minor_Data_Reformatted_df["Infubinol"] , yerr= Infubinol_Error_Bar, label= "Infubinol", marker= "^", color="blue", linestyle='--')
plt.errorbar(x_axis, Minor_Data_Reformatted_df["Ketapril"] , yerr= Ketapril_Error_Bar, label= "Ketapril", marker= "D", color="black", linestyle='--')
plt.errorbar(x_axis, Minor_Data_Reformatted_df["Placebo"] , yerr= Placebo_Error_Bar , label= "Placebo", marker= "s", color="green", linestyle='--')
plt.legend()
plt.title("Tumor Response to Treatment ")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.grid()
# Save the Figure
plt.savefig("Tumor_Response_to_Drugs.png")
```

```python
# Show the Figure
plt.show()
```

![Tumor Response to Treatment](../Images/treatment.png)

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
Mean_Met_Site = combined_mouse_clinical_df.groupby(["Drug", "Timepoint"]).mean()["Metastatic Sites"]
# Convert to DataFrame
Mean_Met_Site = pd.DataFrame(Mean_Met_Site)
# Preview DataFrame
Mean_Met_Site_Index = Mean_Met_Site.reset_index(level=None, drop=False)

Mean_Met_Site_Index.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
Met_Sites_SEM = combined_mouse_clinical_df.groupby(["Drug", "Timepoint"]).sem()["Metastatic Sites"]
# Convert to DataFrame
Met_Sites_SEM = pd.DataFrame(Met_Sites_SEM)
# Preview DataFrame
Met_Sites_SEM_Index = Met_Sites_SEM.reset_index(level=None, drop=False)
Met_Sites_SEM_Index.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
Minor_DataM_Reformatted_df = Mean_Met_Site_Index.pivot_table(index='Timepoint', columns='Drug', values= 'Metastatic Sites')
# Preview that Reformatting worked
Minor_DataM_Reformatted_df.head()
```

```python
# Generate the Plot (with Error Bars)
Capomulin_Error_Bar = Met_Sites_SEM_Index.loc[Met_Sites_SEM_Index["Drug"] == "Capomulin", "Metastatic Sites"]
Infubinol_Error_Bar = Met_Sites_SEM_Index.loc[Met_Sites_SEM_Index["Drug"] == "Infubinol", "Metastatic Sites"]
Ketapril_Error_Bar = Met_Sites_SEM_Index.loc[Met_Sites_SEM_Index["Drug"] == "Ketapril", "Metastatic Sites"]
Placebo_Error_Bar = Met_Sites_SEM_Index.loc[Met_Sites_SEM_Index["Drug"] == "Placebo", "Metastatic Sites"]

#set x axis
x_axis = np.arange(0,50,5)

plt.errorbar(x_axis, Minor_DataM_Reformatted_df["Capomulin"] , yerr= Capomulin_Error_Bar, label= "Capomulin", marker= "o", color="red", linestyle='--')
plt.errorbar(x_axis, Minor_DataM_Reformatted_df["Infubinol"] , yerr= Infubinol_Error_Bar, label= "Infubinol", marker= "^", color="blue", linestyle='--')
plt.errorbar(x_axis, Minor_DataM_Reformatted_df["Ketapril"] , yerr= Ketapril_Error_Bar, label= "Ketapril", marker= "D", color="orange", linestyle='--')
plt.errorbar(x_axis, Minor_DataM_Reformatted_df["Placebo"] , yerr= Placebo_Error_Bar , label= "Placebo", marker= "s", color="green", linestyle='--')
plt.legend()
plt.title("Metastatic Spread During Treatment")
plt.xlabel("Treatment Duration (Days)")
plt.ylabel("Metastatic Sites")
plt.grid()

# Save the Figure
plt.savefig("Tumor_Metsite_Response_to_Drugs.png")

# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
Count_Of_Mice = combined_mouse_clinical_df.groupby(["Drug", "Timepoint"]).count()["Mouse ID"]
# Convert to DataFrame
Count_Of_Mice_df = pd.DataFrame(Count_Of_Mice)

Count_Of_Mice_df_renamed = Count_Of_Mice_df.rename(columns={"Mouse ID": "Mouse Count"})

Count_Of_Mice_df_renamed_Index = Count_Of_Mice_df_renamed.reset_index(level=None, drop=False)

# Preview DataFrame
Count_Of_Mice_df_renamed_Index.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
Minor_DataMice_Reformatted_df = Count_Of_Mice_df_renamed_Index.pivot_table(index='Timepoint', columns='Drug', values= 'Mouse Count')
# Preview the Data Frame
Minor_DataMice_Reformatted_df.head()
```

```python
# Generate the Plot (Accounting for percentages)

x_axis = np.arange(0,50,5)
plt.plot(x_axis, (Minor_DataMice_Reformatted_df["Capomulin"]/25)*100 , label= "Capomulin", marker= "o", color="red", linestyle='dashed')
plt.plot(x_axis, (Minor_DataMice_Reformatted_df["Infubinol"]/25)*100 , label= "Infubinol", marker= "^", color="blue", linestyle='dashed')
plt.plot(x_axis, (Minor_DataMice_Reformatted_df["Ketapril"]/25)*100 , label= "Ketapril", marker= "D", color="orange", linestyle='dashed')
plt.plot(x_axis, (Minor_DataMice_Reformatted_df["Placebo"]/25)*100 , label= "Placebo", marker= "s", color="green", linestyle='dashed')
plt.legend()
plt.title("Survival During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate (%) ")
plt.grid()

# Save the Figure
plt.savefig("Survival_Percentages.png")
# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug

Pct_Change_for_Drug = (Minor_Data_Reformatted_df.iloc[-1] - Minor_Data_Reformatted_df.iloc[0])/Minor_Data_Reformatted_df.iloc[0] * 100

# Display the data to confirm
Pct_Change_for_Drug
```

```python
# Store all Relevant Percent Changes into a Tuple
Percent_Change_Tuple = tuple(zip(Pct_Change_for_Drug.index, Pct_Change_for_Drug))
Percent_Change_Tuple_List = list(Percent_Change_Tuple)
Percent_Change_Tuple_List
```

```python
# Splice the data between passing and failing drugs
Pct_Change_df = pd.DataFrame(Pct_Change_for_Drug)
Pct_Change_df =Pct_Change_df.rename(columns ={0:"Percent Change"})
Pct_Change_df["Pass"] = (Pct_Change_df < 0)
Pct_Change_df

```

```python

# Orient widths. Add labels, tick marks, etc. 
fig, ax = plt.subplots()
y_value = [Pct_Change_for_Drug["Infubinol"], Pct_Change_for_Drug["Ketapril"], Pct_Change_for_Drug["Placebo"]]
x_axis0 = [0]
x_axis1 = [1, 2, 3]


Bar1 = ax.bar(x_axis0, Pct_Change_for_Drug["Capomulin"], color='g', alpha=0.8, align="edge", width = -1)
Bar2 = ax.bar(x_axis1, y_value , color='r', alpha=0.8, align="edge", width = -1)
X_Labels = ["Capomulin", "Infubinol", "Ketapril", "Placebo"]

plt.setp(ax, xticks=[0, 1, 2, 3], xticklabels=["Capomulin", "Infubinol", "Ketapril", "Placebo"],
        yticks=[-20, 0, 20, 40, 60])
ax.set_ylabel('% Tumor Volume Change')
ax.set_title('Tumor Change Over 45 Day Treatment')
ax.grid()

# Use functions to label the percentages of changes
def barslabel(graphics):
    for graph in graphics:
        height = graph.get_height()
        ax.text(graph.get_x() + graph.get_width()/2, .1*height, "%d" %int(height)+ "%",
                ha='center', va='top', color="white")

# Call functions to implement the function calls
barslabel(Bar1)
barslabel(Bar2)
fig.tight_layout()

# Save the Figure
plt.savefig("Tumor_Volume_Change_%.png")

# Show the Figure
fig.show()
```
